```python
print(df.sort_values("math score").head())
import matplotlib.pyplot as plt
df["math score"].hist()
plt.title("MATH SCORE DISTRIBUTION")
plt.xlabel("MARKS")
plt.ylabel("NO. OF STUDENTS")
plt.show()
```

THE GRAPH SHOWS THAT THOUGH THE AVERAGE IS VERY HIGH BUT THE MARKS ARE WIDELY SPREAD IN THE READING SECTION

```python
print(df.sort_values("writing score").head())
df["writing score"].hist()
plt.title("WRITING SCORE DISTRIBUTION")
plt.xlabel("MARKS")
plt.ylabel("NO. OF STUDENTS")
plt.show()
```

THE GRAPH SHOWS THAT A LOT OF STUDENTS HAVE SCORED MORE THAN 80 MARKS AND ALSO THE DATA IS NOT VERY MUCH SPREAD BUT HAS A HIGH AVERAGE
